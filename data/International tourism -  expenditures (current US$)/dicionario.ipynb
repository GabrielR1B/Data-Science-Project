## Dicionário de Dados (11/04)

**Nome:** international_expenditures

**Descrição:** Refere-se aos gastos em viagens ao exterior, incluindo pagamentos a companhias aéreas estrangeiras.  
Pode incluir despesas de viajantes de um dia, exceto quando classificadas separadamente.  
Em alguns países, não considera gastos com transporte de passageiros.  
Os valores estão expressos em dólares americanos correntes U$S.

### Estrutura de dados:

| Coluna        | Tipo     | Descrição                                                                 |
|--------------|----------|---------------------------------------------------------------------------|
| País         | String   | Nome do país                                                              |
| Sigla        | String   | Código do país                                                              |
| Continente   | String   | Continente ao qual o país pertence                                        |
| 2000 a 2020  | Numérico | Valor em U$S gastos em viagens ao exterior                               |

### Observações:
Nesse dataset possui muitos países com poucos dados.Ex: a China só possui 4 anos com dados, mas não foi removida.
