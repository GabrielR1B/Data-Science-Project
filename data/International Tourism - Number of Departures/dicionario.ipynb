## Dicionário de Dados (11/04)

**Nome:** international_tourism_number_departures

**Descrição:** Turistas internacionais emissivos são as viagens que pessoas fazem para fora do seu país (não para trabalhar).
O dado conta quantas saídas aconteceram, e não quantas pessoas viajaram.
Ou seja, se a mesma pessoa viajar várias vezes, cada viagem é contada separadamente.
### Estrutura de dados:

| Coluna        | Tipo     | Descrição                                                                 |
|--------------|----------|---------------------------------------------------------------------------|
| País         | String   | Nome do país                                                              |
| Sigla        | String   | Código do país                                                              |
| Continente   | String   | Continente ao qual o país pertence                                        |
| 1995 a 2018  | Numérico | Número de saídas                           |

### Observações:
Esse dataset possui muitos países com pouco ou nenhum dado que foram removidos (mais de 100). Mas nenhum dos países removidos é muito importante (consultar planilha dos removidos)
