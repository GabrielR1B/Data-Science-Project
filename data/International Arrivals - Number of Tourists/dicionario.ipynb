## Dicionário de Dados (11/04)

**Nome:** international_arrivals_number_tourists

**Descrição:** Esta tabela descreve o formato e o significado das colunas presentes no conjunto de dados, que contém informações sobre e o número de chegadas de turistas internacionais ao longo dos anos em cada país.

### Estrutura de dados:

| Coluna        | Tipo     | Descrição                                                                 |
|--------------|----------|---------------------------------------------------------------------------|
| País         | String   | Nome do país                                                              |
| Sigla        | String   | Código do país                                                              |
| Continente   | String   | Continente ao qual o país pertence                                        |
| 2004 a 2018  | Numérico | Número total de chegadas de turistas internacionais em cada ano      |
