## Dicionário de Dados (11/04)

**Nome:** international_tourism_receipts

**Descrição:** Refere-se às receitas geradas pelos gastos de visitantes internacionais no país, incluindo pagamentos a companhias aéreas nacionais.  
Abrange também valores pagos antecipadamente por bens e serviços no destino.  
Pode incluir visitantes de um dia, exceto quando classificados separadamente.  
Os dados estão em dólares americanos correntes U$S.
### Estrutura de dados:

| Coluna        | Tipo     | Descrição                                                                 |
|--------------|----------|---------------------------------------------------------------------------|
| País         | String   | Nome do país                                                              |
| Sigla        | String   | Código do país                                                              |
| Continente   | String   | Continente ao qual o país pertence                                        |
| 2004 a 2018  | Numérico | Receita gerada pelos visitantes em U$S                           |

### Observações:
Nesse dataset possui muitos países com poucos dados que foram removidos.Ex: united kingdom e spain só possui 3 anos com dados no periodo.
